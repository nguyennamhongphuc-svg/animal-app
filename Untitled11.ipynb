```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
import urllib.request
import os

# 1. Tải dữ liệu 10 loài động vật từ Google QuickDraw
classes = ['cat', 'dog', 'pig', 'mouse', 'elephant', 'bird', 'fish', 'horse', 'rabbit', 'snake']
base_url = 'https://storage.googleapis.com/quickdraw_dataset/full/numpy_bitmap/'

X, y = [], []
for idx, class_name in enumerate(classes):
    path = f'{class_name}.npy'
    if not os.path.exists(path):
        urllib.request.urlretrieve(base_url + class_name + '.npy', path)
    data = np.load(path)[:5000] # Lấy 5000 mẫu mỗi loại
    X.append(data)
    y.append(np.full(data.shape[0], idx))

X = np.concatenate(X).reshape(-1, 28, 28, 1).astype('float32') / 255.0
y = np.concatenate(y)

# 2. Kiến trúc CNN - Hiệu quả vượt trội cho hình ảnh
#

model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(len(classes), activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=10, validation_split=0.2)

# 3. Lưu model
model.save('animal_model.h5')
print("Đã tạo file animal_model.h5 thành công!")
```
